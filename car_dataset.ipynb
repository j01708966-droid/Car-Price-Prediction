```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import drive
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
import joblib
```

```python
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/car dataset.csv'
df = pd.read_csv(file_path)
print(df)
```

```python
print(df.head())
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df['max_power'] = pd.to_numeric(df['max_power'], errors='coerce')
```

```python
numeric_col = ['mileage(km/ltr/kg)', 'engine(cc)', 'max_power', 'seats']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

categorical_col = df.select_dtypes(include='object').columns
for col in categorical_col:
  df[col] = le.fit_transform(df[col].astype(str))
```

```python
x = df.drop('selling_price', axis=1)
y = df['selling_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
sample_data = x.iloc[0:1]
predicted_price = rf.predict(sample_data)
print("Predicted Selling Price:", predicted_price[0])
```

```python
joblib.dump(rf, "car_price_prediction_model.pkl")
print("Model Saved Successfully")
```
